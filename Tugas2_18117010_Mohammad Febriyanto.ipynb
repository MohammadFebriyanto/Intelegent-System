Berikut ini merupakan contoh digits recognition. Data yang digunakan merupakan dataset yang tersedia di scikit-learn. Kemudian hasil pemodelan dari data tersebut diujicobakan dengan data tulisan tangan peserta mata kuliah sistem cerdas.

```python
from sklearn import datasets
from scipy import misc
```

```python
digits = datasets.load_digits()
features = digits.data 
labels = digits.target
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(features, labels, stratify=labels, test_size=0.2)
```

```python
from sklearn.svm import SVC
from sklearn import metrics

clf = SVC(degree=4).fit(X_train, y_train)
predicted = clf.predict(X_test)
print("SVC Accuracy : {}".format(metrics.accuracy_score(y_test, predicted)))
```

```python
count = 0
right = 0

for i in range(0,10):    
    for k in range(0,31):
        img = misc.imread('./testing/'+str(i)+'/'+str(k)+'.jpg')
        img = misc.imresize(img, (8,8))
        img = img.astype(digits.images.dtype)
        img = misc.bytescale(img, high=16, low=0)
        count += 1

        
        try:
            x_test = []
            for eachRow in img:
                for eachPixel in eachRow:
                    x_test.append(sum(eachPixel)/3.0)
            if clf.predict([x_test])[0] == i:
                right += 1
        except TypeError : #Menghindari data rusak
            count += -1

print('accuracy_score = %.5f ' %(right*100/count) +'%' )
```
