# KNN

```python
import operator
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error, r2_score
```

```python
test_filename="II4035-regresi-test.csv"
train_filename="II4035-regresi-train.csv"

training=np.genfromtxt(train_filename,delimiter=',',skip_header=1)
test    =np.genfromtxt(test_filename,delimiter=',',skip_header=1)
```

```python
train_x=training[:,1]
train_y=training[:,2]
test = test[:,1]
```

```python
# plot dengan matplotlib. manual: https://matplotlib.org/api/_as_gen/matplotlib.pyplot.plot.html
plt.plot(train_x, train_y, "bx") # blue line
plt.xlabel("$x_1$", fontsize=18)
plt.ylabel("$y$", rotation=0, fontsize=18)
plt.axis([0, 2, -1, 1])
plt.show()
```

```python
x = train_x[:, np.newaxis]
y = train_y[:, np.newaxis]
test = test[:, np.newaxis]
```

```python
model = KNeighborsRegressor()
model.fit(x, y)
y_pred = model.predict(x)
```

```python
rmse = np.sqrt(mean_squared_error(y,y_pred))
r2 = r2_score(y,y_pred)
print(rmse)
print(r2)
```

```python
plt.scatter(x, y, s=10)
# sort the values of x before line plot
sort_axis = operator.itemgetter(0)
sorted_zip = sorted(zip(x,y_pred), key=sort_axis)
x, y_poly_pred = zip(*sorted_zip)
plt.plot(x, y_pred, color='m')
plt.show()
```

```python
# Underfit

model = KNeighborsRegressor(n_neighbors=50)
model.fit(x, y)
y_poly_pred = model.predict(x)
    
plt.scatter(x, y, s=10)
# sort the values of x before line plot
sort_axis = operator.itemgetter(0)
sorted_zip = sorted(zip(x,y_poly_pred), key=sort_axis)
x, y_poly_pred = zip(*sorted_zip)
plt.plot(x, y_poly_pred, color='m')
plt.show()
```

```python
# overfit

model = KNeighborsRegressor(n_neighbors=1)
model.fit(x, y)
y_poly_pred = model.predict(x)
    
plt.scatter(x, y, s=10)
# sort the values of x before line plot
sort_axis = operator.itemgetter(0)
sorted_zip = sorted(zip(x,y_poly_pred), key=sort_axis)
x, y_poly_pred = zip(*sorted_zip)
plt.plot(x, y_poly_pred, color='m')
plt.show()
```

```python
# Optimal


model = KNeighborsRegressor(n_neighbors=6)
model.fit(x, y)
y_poly_pred = model.predict(x)
    
plt.scatter(x, y, s=10)
# sort the values of x before line plot
sort_axis = operator.itemgetter(0)
sorted_zip = sorted(zip(x,y_poly_pred), key=sort_axis)
x, y_poly_pred = zip(*sorted_zip)
plt.plot(x, y_poly_pred, color='m')
plt.show()
```

```python
predicted = model.predict(test)

from numpy import asarray
from numpy import savetxt
# define data
data = predicted
# save to csv file
savetxt('hasilprediksi_KNN.csv', data, delimiter=',')
```
