```python
import operator
import numpy as np
import matplotlib.pyplot as plt

from sklearn.tree import DecisionTreeRegressor
```

```python
test_filename="II4035-regresi-test.csv"
train_filename="II4035-regresi-train.csv"

training=np.genfromtxt(train_filename,delimiter=',',skip_header=1)
test    =np.genfromtxt(test_filename,delimiter=',',skip_header=1)
```

```python
train_x=training[:,1]
train_y=training[:,2]
test = test[:,1]
```

```python
x = train_x[:, np.newaxis]
y = train_y[:, np.newaxis]
test = test[:, np.newaxis]
```

```python
#Underfit

model = DecisionTreeRegressor(max_depth=1)
model.fit(x, y)
y_poly_pred = model.predict(x)
    
plt.scatter(x, y, s=30)
# sort the values of x before line plot
sort_axis = operator.itemgetter(0)
sorted_zip = sorted(zip(x,y_poly_pred), key=sort_axis)
x, y_poly_pred = zip(*sorted_zip)
plt.plot(x, y_poly_pred, color='m')
plt.show()
```

```python
#overfit

model = DecisionTreeRegressor(max_depth=100)
model.fit(x, y)
y_poly_pred = model.predict(x)
    
plt.scatter(x, y, s=30)
# sort the values of x before line plot
sort_axis = operator.itemgetter(0)
sorted_zip = sorted(zip(x,y_poly_pred), key=sort_axis)
x, y_poly_pred = zip(*sorted_zip)
plt.plot(x, y_poly_pred, color='m')
plt.show()
```

```python
#GOOD MODEL

model = DecisionTreeRegressor(max_depth=4)
model.fit(x, y)
y_poly_pred = model.predict(x)
    
plt.scatter(x, y, s=30)
# sort the values of x before line plot
sort_axis = operator.itemgetter(0)
sorted_zip = sorted(zip(x,y_poly_pred), key=sort_axis)
x, y_poly_pred = zip(*sorted_zip)
plt.plot(x, y_poly_pred, color='m')
plt.show()
```

```python
predicted = model.predict(test)

from numpy import asarray
from numpy import savetxt
# define data
data = predicted
# save to csv file
savetxt('hasilprediksi_DecisionTreeRegressor.csv', data, delimiter=',')
```
